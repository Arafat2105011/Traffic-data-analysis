```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(pol['driver_age'], bins=26)
plt.title('Distribution of Driver Age')
plt.show()
```

```python
sns.countplot(x='driver_gender', data=pol)
plt.title('Driver Gender Count')
plt.show()
```

```python
pol['stop_datetime'] = pd.to_datetime(pol['stop_date'] + ' ' + pol['stop_time'])


daily_violations = pol.set_index('stop_datetime')['violation'].resample('D').count()


daily_violations.plot(figsize=(17, 8))
plt.title('Daily Violations')
plt.ylabel('Count')
plt.show()
```
